```python
plt.figure(figsize=(10,15))
for i,column in enumerate(cat):
    plt.subplot(2,2,i+1)
    bdf[column].value_counts()
    property=bdf[column].value_counts(normalize=True).keys()
    count=bdf[column].value_counts(normalize=True).values
    Data=pd.DataFrame(zip(property,count),columns=[column,count])
    n=bdf[column].nunique()
    l=[0.1 for i in range(n)]
    plt.title(column)
    plt.pie(x=count,labels=property,autopct='%0.2f%%',shadow=True,startangle=90,explode=l)
plt.show()
```

**Machine Learning**

```python
# seperate input and output column

X=bdf.drop('Purchase',axis=1)

y=bdf['Purchase']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1234)
```

```python
print(bdf.shape)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

**LInear Regression**

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_prediction=lr.predict(X_test)
```

```python
len(y_prediction)
```

```python
len(y_test)
```

```python
y_test[:5].values,y_prediction[:5]
```

```python
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
```

```python
mean_square_error1=mean_squared_error(y_test,y_prediction)
mean_absolute_error1=mean_absolute_error(y_test,y_prediction)
R2_Score1=r2_score(y_test,y_prediction)
```

```python
print("mean square error1",mean_square_error1)
print("mean absolute error1",mean_absolute_error1)
print("R2 Score1",R2_Score1)
```

**Decision tree**

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
dt.fit(X_train,y_train)
```

```python
y_prediction=dt.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
```

```python
mean_square_error2=mean_squared_error(y_test,y_prediction)
mean_absolute_error2=mean_absolute_error(y_test,y_prediction)
R2_Score2=r2_score(y_test,y_prediction)
```

```python
print("mean square error2",mean_square_error2)
print("mean absolute error2",mean_absolute_error2)
print("R2 Score2",R2_Score2)
```

**Random forest**

```python

from sklearn.ensemble import RandomForestRegressor
Rf=RandomForestRegressor()
Rf.fit(X_train,y_train)
```

```python
y_prediction=Rf.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
```

```python
mean_square_error3=mean_squared_error(y_test,y_prediction)
mean_absolute_error3=mean_absolute_error(y_test,y_prediction)
R2_Score3=r2_score(y_test,y_prediction)
```

```python
print("mean square error3",mean_square_error3)
print("mean absolute error3",mean_absolute_error3)
print("R2 Score3",R2_Score3)
```
